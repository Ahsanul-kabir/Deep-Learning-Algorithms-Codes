```python
data_set = pd.read_csv('iris.csv')
data_set
```

```python
real_x = data_set.iloc[:,0:4].values # IV
real_y = data_set.iloc[:,4].values # DV/op
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y1 = encoder.fit_transform(real_y)
Y = pd.get_dummies(y1).values

from sklearn.model_selection import train_test_split # aage eta use hotw-->from sklearn.cross_validation import train_test_split
training_x,test_x,training_y,test_y = train_test_split(real_x,Y,test_size = 0.2, random_state=0)
```

```python
y1
```

```python
Y
```

```python
model = Sequential()
model.add(Dense(4, input_shape = (4,), activation='relu'))

model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(training_x, training_y, batch_size=10, nb_epoch=100)
```

```python
# predicting the test result
y_pred = model.predict(test_x)
test_y_class = np.argmax(test_y,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
# making the confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(test_y_class,y_pred_class))
print(classification_report(test_y_class,y_pred_class))
```
